```python
# prep: import modules and get pwd
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import getpass  # To get the password without showing the input
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine=create_engine(connection_string)
```

```python
query = '''
SELECT
    f.title AS unique_film,
    f.rental_rate,
    f.rating,
    f.length,
    c.name AS film_category,
    COUNT(r.rental_id) AS total_rentals,
    MAX(CASE WHEN YEAR(r.rental_date) = 2005 AND MONTH (r.rental_date) = 05
		THEN True
        ELSE False
	END) AS "rented_in_may"
FROM film AS f
LEFT JOIN inventory AS i
ON f.film_id = i.film_id
LEFT JOIN rental AS r
ON i.inventory_id = r.inventory_id
JOIN film_category as fc
ON f.film_id = fc.film_id
JOIN category AS c
ON c.category_id = fc.category_id
GROUP BY unique_film, f.rental_rate, f.rating, f.length, film_category;
'''

df = pd.read_sql_query(query, engine)

df
```

```python
#Checking numerical columns for correlation and multicollinearity
import matplotlib.pyplot as plt
import seaborn as sns
corr_matrix=df.select_dtypes(np.number).corr(method='pearson')  # default
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#X/y Split (since I only found a significant correlation in the total rentals category I decided to only keep that numerical column)

X=df.drop(["rented_in_may", "length", "rental_rate"], axis=1)
y=df["rented_in_may"]
```

```python
#Train/Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
#Selecting numerical columns
X_train_num = X_train.select_dtypes(include = np.number)
X_test_num  = X_test.select_dtypes(include = np.number)
```

```python
#applying MinMaxScaler to numerical columns
from sklearn.preprocessing import MinMaxScaler

#fitting the train set
MinMaxScaler = MinMaxScaler()
transformer = MinMaxScaler.fit(X_train_num) # need to keep transformer

#Saving scaler
import pickle
import os

path = "scalers/"

isExist = os.path.exists(path)
if not isExist:
  os.makedirs(path)
  print("The new directory is created!")

filename = "MinMaxScaler.pkl"
with open(path+filename, "wb") as file:
    pickle.dump(MinMaxScaler, file)
```

```python
#transforming the train and test sets
X_train_normalized = transformer.transform(X_train_num)
X_test_normalized  = transformer.transform(X_test_num)

#transforming the arrays into dataframes
X_train_norm = pd.DataFrame(X_train_normalized, columns=X_train_num.columns, index= X_train_num.index)
X_test_norm  = pd.DataFrame(X_test_normalized, columns=X_test_num.columns, index= X_test_num.index)
```

```python
display(X_train_norm.shape)
display(X_test_norm.shape)
X_train_norm.describe().T
```

```python
#Selecting categorical columns (droping "unique_film" column because that's the unique identifier of each row)
X_train_cat = X_train.select_dtypes('object')
X_test_cat  = X_test.select_dtypes('object')
X_train_cat.drop("unique_film", axis=1, inplace=True)
X_test_cat.drop("unique_film", axis=1, inplace=True)
```

```python
X_train_cat
```

```python
# applying OneHotEncoder to the categorical columns
from sklearn.preprocessing import OneHotEncoder

#fitting on the train set
encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_cat)

#saving encoder
path = "encoders/"

isExist = os.path.exists(path)
if not isExist:
  os.makedirs(path)
  print("The new directory is created!")

filename = "OneHotEncoder.pkl"
with open(path+filename, "wb") as file:
   pickle.dump(encoder, file)
```

```python
#transforming both the train and test
X_train_cat_np = encoder.transform(X_train_cat).toarray()
X_test_cat_np  = encoder.transform(X_test_cat).toarray()

#converting to dataframes
X_train_cat = pd.DataFrame(X_train_cat_np, columns=encoder.get_feature_names_out(), 
                           index=X_train_cat.index)
X_test_cat = pd.DataFrame(X_test_cat_np,  columns=encoder.get_feature_names_out(),
                          index=X_test_cat.index)
```

```python
X_train_cat
```

```python
X_test_cat
```

```python
#building X_train
X_train_transformed = pd.concat([X_train_norm, X_train_cat], axis=1)
```

```python
X_train_transformed
```

```python
#building X_test
X_test_transformed = pd.concat([X_test_norm, X_test_cat], axis=1)
```

```python
X_test_transformed
```

4. Create a logistic regression model to predict 'rented_in_may' from the cleaned data.

```python
#creating logistic regression model
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')

classification.fit(X_train_transformed, y_train)

#saving model
path = "models/"
    
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    print("The new directory is created!")

filename = "LogisticRegression.pkl" 
with open(path+filename, "wb") as file:
    pickle.dump(classification, file)
```

```python
#accuracy score
y_test_pred = classification.predict(X_test_transformed)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_test_pred))
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_test_pred))
```

```python
#analyzing the confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

confusion_matrix(y_test, y_test_pred)
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)

disp = ConfusionMatrixDisplay(cm_test,display_labels=classification.classes_)
disp.plot()
plt.show()
```

```python
#analyzing the kappa score
from sklearn.metrics import cohen_kappa_score

cohen_kappa_score(y_test, y_test_pred)
```
